```python
train_data.drop(columns=['is_fraud','gender'],inplace=True)
test_data.drop(columns=['is_fraud','gender'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

x_train['trans_num_encoded'] = label_encoder.fit_transform(x_train['trans_num'])
x_test['trans_num_encoded'] = label_encoder.fit_transform(x_test['trans_num'])
x_train.drop('trans_num', axis=1, inplace=True)
x_test.drop('trans_num', axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=1)
```
